```python
from sklearn.model_selection import train_test_split
import json
import warnings
import numpy as np
import pandas as pd

from sklearn.impute import SimpleImputer
from sklearn.metrics import mean_squared_log_error, roc_auc_score
from sklearn.model_selection import ParameterSampler, train_test_split
from sklearn.tree import DecisionTreeRegressor, DecisionTreeClassifier
from sklearn.ensemble import (
    RandomForestRegressor,
    RandomForestClassifier,
    HistGradientBoostingRegressor,
    HistGradientBoostingClassifier,
)

import lightgbm as lgb
import xgboost as xgb
from catboost import CatBoostRegressor, CatBoostClassifier

warnings.filterwarnings("ignore")
pd.options.display.max_rows = 1000

dataset = pd.read_csv(f"1_{NAME}.csv")

if dataset.columns[0].startswith("Unnamed"):
    dataset = dataset.drop(columns=dataset.columns[0])

dataset = dataset.reset_index(drop=True)

SHAPE = dataset.shape

train_mask = dataset["dtype"] == "train"
test_mask = dataset["dtype"] == "test"

train_all = dataset.loc[train_mask].copy().reset_index(drop=True)
test = dataset.loc[test_mask].copy().reset_index(drop=True)

print("dataset:", dataset.shape)
print("train:", train_all.shape)
print("test:", test.shape)


# Feature block
target_cols = ["__price", "__churn", "dtype"]
object_cols = dataset.select_dtypes(include=["object"]).columns.tolist()

feature_cols_all = [
    c for c in dataset.columns
    if c not in target_cols and c not in object_cols
]

X_numeric = dataset[feature_cols_all].replace([np.inf, -np.inf], np.nan)
train_numeric = X_numeric.loc[train_mask].copy().reset_index(drop=True)
test_numeric = X_numeric.loc[test_mask].copy().reset_index(drop=True)

print("numeric features:", len(feature_cols_all))


# OOT split for price
if "timestamp" in train_all.columns:
    train_all["timestamp_dt"] = pd.to_datetime(train_all["timestamp"])
    train_sorted = train_all.sort_values("timestamp_dt").reset_index(drop=True)

    oot_cutoff = train_sorted["timestamp_dt"].quantile(0.80)

    dev_mask_price = train_sorted["timestamp_dt"] < oot_cutoff
    oot_mask_price = train_sorted["timestamp_dt"] >= oot_cutoff
else:
    split_pos = int(len(train_all) * 0.8)
    train_sorted = train_all.copy().reset_index(drop=True)
    dev_mask_price = np.arange(len(train_sorted)) < split_pos
    oot_mask_price = np.arange(len(train_sorted)) >= split_pos

train_numeric_reset = train_numeric.reset_index(drop=True)
train_numeric_sorted = train_numeric_reset.loc[train_sorted.index].reset_index(drop=True)

X_dev_full = train_numeric_sorted.loc[dev_mask_price].reset_index(drop=True)
X_oot_full = train_numeric_sorted.loc[oot_mask_price].reset_index(drop=True)

y_price_dev = train_sorted.loc[dev_mask_price, "__price"].reset_index(drop=True)
y_price_oot = train_sorted.loc[oot_mask_price, "__price"].reset_index(drop=True)

print("price dev:", X_dev_full.shape, "price oot:", X_oot_full.shape)

# Separate split for churn
train_all_reset = train_all.reset_index(drop=True)
train_numeric_reset = train_numeric.reset_index(drop=True)

X_churn_source = train_numeric_reset.copy()
y_churn_source = train_all_reset["__churn"].astype(int).copy()

X_churn_dev_full, X_churn_oot_full, y_churn_dev, y_churn_oot = train_test_split(
    X_churn_source,
    y_churn_source,
    test_size=0.20,
    random_state=47,
    stratify=y_churn_source
)

X_churn_dev_full = X_churn_dev_full.reset_index(drop=True)
X_churn_oot_full = X_churn_oot_full.reset_index(drop=True)
y_churn_dev = y_churn_dev.reset_index(drop=True)
y_churn_oot = y_churn_oot.reset_index(drop=True)

print("churn dev:", X_churn_dev_full.shape, "churn oot:", X_churn_oot_full.shape)


# Raw metrics for model selection
def metric_for_price_raw(y_true, y_pred):
    y_pred = np.clip(np.asarray(y_pred), 1e-9, None)
    return -mean_squared_log_error(y_true=y_true, y_pred=y_pred)

def metric_for_churn_raw(y_true, y_score):
    return roc_auc_score(y_true=y_true, y_score=y_score)


# Feature preselection helper

def select_top_model_features(X_dev, y_dev, task="regression", top_k=150):
    imputer = SimpleImputer(strategy="median")

    X_dev_imp = pd.DataFrame(
        imputer.fit_transform(X_dev),
        index=X_dev.index,
        columns=X_dev.columns
    )

    if task == "regression":
        selector = lgb.LGBMRegressor(
            n_estimators=200,
            learning_rate=0.05,
            num_leaves=31,
            random_state=47,
            n_jobs=-1,
            verbosity=-1
        )
        selector.fit(X_dev_imp, y_dev)
    else:
        selector = lgb.LGBMClassifier(
            n_estimators=200,
            learning_rate=0.05,
            num_leaves=31,
            random_state=47,
            n_jobs=-1,
            verbosity=-1
        )
        selector.fit(X_dev_imp, y_dev)

    imp = pd.Series(selector.feature_importances_, index=X_dev_imp.columns).sort_values(ascending=False)

    selected = imp[imp > 0].index.tolist()
    if len(selected) > top_k:
        selected = selected[:top_k]
    if len(selected) < min(top_k, len(imp)):
        selected = imp.head(top_k).index.tolist()

    return selected

price_features = select_top_model_features(
    X_dev_full, y_price_dev,
    task="regression",
    top_k=160
)

churn_features = select_top_model_features(
    X_churn_dev_full, y_churn_dev,
    task="classification",
    top_k=180
)

print("price_features:", len(price_features))
print("churn_features:", len(churn_features))


# Final imputers

price_imputer = SimpleImputer(strategy="median")
X_price_dev = price_imputer.fit_transform(X_dev_full[price_features]).astype("float32")
X_price_oot = price_imputer.transform(X_oot_full[price_features]).astype("float32")

churn_imputer = SimpleImputer(strategy="median")
X_churn_dev = churn_imputer.fit_transform(X_churn_dev_full[churn_features]).astype("float32")
X_churn_oot = churn_imputer.transform(X_churn_oot_full[churn_features]).astype("float32")


# Random seed for search

rng = 42

def build_price_model(alg, params):
    if alg == "Decision Tree":
        return DecisionTreeRegressor(random_state=47, **params)

    if alg == "Random Forest":
        return RandomForestRegressor(random_state=47, n_jobs=-1, **params)

    if alg == "Hist Gradient Boosting":
        return HistGradientBoostingRegressor(
            random_state=47,
            early_stopping=True,
            validation_fraction=0.1,
            n_iter_no_change=20,
            **params
        )

    if alg == "LightGBM":
        return lgb.LGBMRegressor(
            random_state=47,
            objective="regression",
            n_jobs=-1,
            verbosity=-1,
            **params
        )

    if alg == "XGBoost":
        return xgb.XGBRegressor(
            random_state=47,
            objective="reg:squarederror",
            tree_method="hist",
            n_jobs=-1,
            eval_metric="rmse",
            **params
        )

    if alg == "CatBoost":
        return CatBoostRegressor(
            random_seed=47,
            loss_function="RMSE",
            verbose=False,
            **params
        )

    raise ValueError(f"Unknown algorithm: {alg}")

def build_churn_model(alg, params):
    if alg == "Decision Tree":
        return DecisionTreeClassifier(random_state=47, **params)

    if alg == "Random Forest":
        return RandomForestClassifier(random_state=47, n_jobs=-1, **params)

    if alg == "Hist Gradient Boosting":
        return HistGradientBoostingClassifier(
            random_state=47,
            early_stopping=True,
            validation_fraction=0.1,
            n_iter_no_change=20,
            **params
        )

    if alg == "LightGBM":
        return lgb.LGBMClassifier(
            random_state=47,
            objective="binary",
            n_jobs=-1,
            verbosity=-1,
            **params
        )

    if alg == "XGBoost":
        return xgb.XGBClassifier(
            random_state=47,
            objective="binary:logistic",
            tree_method="hist",
            n_jobs=-1,
            eval_metric="auc",
            **params
        )

    if alg == "CatBoost":
        return CatBoostClassifier(
            random_seed=47,
            loss_function="Logloss",
            eval_metric="AUC",
            verbose=False,
            **params
        )

    raise ValueError(f"Unknown algorithm: {alg}")

def fit_price_model(model, alg, X_dev, y_dev, X_oot, y_oot):
    if alg == "LightGBM":
        model.fit(
            X_dev, y_dev,
            eval_set=[(X_oot, y_oot)],
            callbacks=[lgb.early_stopping(30, verbose=False)]
        )
    elif alg == "CatBoost":
        model.fit(
            X_dev, y_dev,
            eval_set=(X_oot, y_oot),
            use_best_model=True,
            verbose=False
        )
    else:
        model.fit(X_dev, y_dev)
    return model

def fit_churn_model(model, alg, X_dev, y_dev, X_oot, y_oot):
    if alg == "LightGBM":
        model.fit(
            X_dev, y_dev,
            eval_set=[(X_oot, y_oot)],
            callbacks=[lgb.early_stopping(30, verbose=False)]
        )
    elif alg == "CatBoost":
        model.fit(
            X_dev, y_dev,
            eval_set=(X_oot, y_oot),
            use_best_model=True,
            verbose=False
        )
    else:
        model.fit(X_dev, y_dev)
    return model
```

```python
rng = 42
n_experiments = 11  

price_spaces = {
    "Decision Tree": {
        "max_depth": [4, 6, 8, 10, 12, None],
        "min_samples_leaf": [5, 10, 20, 40, 80],
        "min_samples_split": [10, 20, 50, 100, 200],
        "criterion": ["squared_error", "friedman_mse"],
    },

    "Random Forest": {
        "n_estimators": [80, 120, 160],
        "max_depth": [8, 12, 16, 20, None],
        "min_samples_leaf": [1, 2, 5, 10],
        "max_features": ["sqrt", 0.5, 0.8],
    },

    "Hist Gradient Boosting": {
        "learning_rate": [0.03, 0.05, 0.08],
        "max_iter": [150, 220, 300],
        "max_depth": [4, 6, 8, None],
        "min_samples_leaf": [10, 20, 40, 80],
        "l2_regularization": [0.0, 1.0, 3.0, 5.0],
        "max_leaf_nodes": [15, 31, 63],
    },

    "LightGBM": {
        "learning_rate": [0.03, 0.05, 0.08],
        "n_estimators": [200, 320, 450],
        "num_leaves": [31, 63, 127],
        "min_child_samples": [10, 20, 40, 80],
        "subsample": [0.7, 0.85, 1.0],
        "colsample_bytree": [0.6, 0.8, 1.0],
        "reg_lambda": [0.0, 1.0, 3.0, 5.0],
    },

    "XGBoost": {
        "learning_rate": [0.03, 0.05, 0.08],
        "n_estimators": [200, 320, 450],
        "max_depth": [4, 6, 8],
        "min_child_weight": [1, 3, 5],
        "subsample": [0.7, 0.85, 1.0],
        "colsample_bytree": [0.6, 0.8, 1.0],
        "reg_lambda": [1.0, 3.0, 5.0, 10.0],
        "gamma": [0.0, 0.1, 0.3],
    },

    "CatBoost": {
        "learning_rate": [0.03, 0.05, 0.08],
        "iterations": [200, 320, 450],
        "depth": [4, 6, 8, 10],
        "l2_leaf_reg": [1, 3, 5, 7, 10],
        "bagging_temperature": [0.0, 0.5, 1.0],
        "random_strength": [0.5, 1.0, 2.0],
    },
}

scoring_price_rows = []
best_price = None

for alg, space in price_spaces.items():
    sampled_params = list(ParameterSampler(space, n_iter=n_experiments, random_state=rng))

    for params in sampled_params:
        model = build_price_model(alg, params)
        model = fit_price_model(model, alg, X_price_dev, y_price_dev, X_price_oot, y_price_oot)

        pred_dev = model.predict(X_price_dev)
        pred_oot = model.predict(X_price_oot)

        score_train = metric_for_price(y_price_dev, np.clip(pred_dev, 1e-9, None))
        score_test = metric_for_price(y_price_oot, np.clip(pred_oot, 1e-9, None))

        row = {
            "alg": alg,
            "hyperparametes": json.dumps(params, sort_keys=True),
            "metric_name": "metric_for_price",
            "score_train": score_train,
            "score_test": score_test,
        }
        scoring_price_rows.append(row)

        raw_test = metric_for_price_raw(y_price_oot, pred_oot)

        if (best_price is None) or (raw_test > best_price["raw_score_test"]):
            best_price = {
                "alg": alg,
                "params": params,
                "raw_score_test": raw_test,
                "score_train": score_train,
                "score_test": score_test,
            }

scoring_price = pd.DataFrame(scoring_price_rows).sort_values(
    ["score_test", "score_train"],
    ascending=False
).reset_index(drop=True)

print("BEST REGRESSION MODEL")
print(best_price["alg"])
print(best_price["params"])
print("train:", best_price["score_train"], "test:", best_price["score_test"])

scoring_price.head(20)
```
